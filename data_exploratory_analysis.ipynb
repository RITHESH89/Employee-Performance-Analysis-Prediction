```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_excel("employee_data.xls")
```

```python
sns.countplot(x="EducationBackground", data=df)
plt.title("Education Background Distribution")
plt.show()
```

```python
print(df.columns)
```

```python
if "Department" in df.columns:
    df.groupby("Department")["PerformanceRating"].mean().plot(kind='bar')
    plt.title("Department-wise Performance")
    plt.show()
```

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
