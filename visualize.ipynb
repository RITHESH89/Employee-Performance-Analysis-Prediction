```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

```python
df = pd.read_excel("employee_data.xls")
model = joblib.load("/content/model.pkl")
```

```python
# Feature Importance
importance = pd.Series(model.feature_importances_, index=df.drop("PerformanceRating", axis=1).columns)
```

```python
importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```
